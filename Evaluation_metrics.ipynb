This notebook is used for the evaluation of the chosen metrics for the stylized images we produced with all the architectures.

```python
# Initialize variables to store data
data_efficient = []
data_mobile = []
data_vgg = []

# directories for the three architectures
filename = ['./StylizedImages/Efficient/Metrics.txt','./StylizedImages/Mobile/Metrics.txt','./StylizedImages/Mobile/Metrics.txt']

filename_efficient = './StylizedImages/Efficient/Metrics.txt'
filename_mobile = './StylizedImages/Mobile/Metrics.txt'
filename_vgg = './StylizedImages/VGG/Metrics.txt'
# Open the text file for reading  
with open(filename_efficient, 'r') as file:
    entry = {}
    for line in file:
        line = line.strip()
        if line.startswith('Image and Style:'):
            entry['Image'] = line.split(':')[1].strip()
        elif line.startswith('SSIM Value:'):
            entry['SSIM Value'] = float(line.split(':')[1].strip())
        else:
            entry['Values'] = [float(value) for value in line.split()]
            data_efficient.append(entry)
            entry = {}
# Create a new list to store the extracted strings
image_strings = []

# Iterate through the original list
for element in data_efficient:
    if 'Image' in element:
        image_strings.append(element['Image'])


# Initialize a new list to store grouped elements
grouped_list_efficient = []

# Iterate through the input list
for i in range(0, len(data_efficient), 3):
    group = {}
    for j in range(3):
        if 'Image' in data_efficient[i + j]:
            group['Image'] = data_efficient[i + j]['Image']
            group['SSIM Value'] = data_efficient[i + j]['SSIM Value']
        if 'Values' in data_efficient[i + j]:
            if 'Values' not in group:
                group['Values'] = []
            group['Values'].extend(data_efficient[i + j]['Values'])
    grouped_list_efficient.append(group)

# Calculate averages of chi score
chi_values_list_efficient = []
for i in range (len(grouped_list_efficient)):
    #chi_values.append(grouped_list[i]['Values'])
    average = sum(grouped_list_efficient[i]['Values']) / len(grouped_list_efficient[i]['Values'])
    #print("Average Chi Value:", average)
    # Normalization of chi score
    average = 1/(1+average)
    #print("Normalized Average Chi Value:", average)
    chi_values_list_efficient.append(average)

# Iterate through the list
ssim_value_list_efficient = []
for item in data_efficient:
    if 'SSIM Value' in item:
        ssim_value = item['SSIM Value']
        
        # Print or use the extracted data as needed
        #print("SSIM Value:", ssim_value)
        ssim_value_list_efficient.append(ssim_value)

# Calculate an average combined score
average_score_efficient = []
for i in range(len(chi_values_list_efficient)):
    average_score_efficient.append( (chi_values_list_efficient[i] + ssim_value_list_efficient[i])/2);
    #print(average_score_efficient[i])
    






# Open the text file for reading  
with open(filename_mobile, 'r') as file:
    entry = {}
    for line in file:
        line = line.strip()
        if line.startswith('Image and Style:'):
            entry['Image'] = line.split(':')[1].strip()
        elif line.startswith('SSIM Value:'):
            entry['SSIM Value'] = float(line.split(':')[1].strip())
        else:
            entry['Values'] = [float(value) for value in line.split()]
            data_mobile.append(entry)
            entry = {}

# Initialize a new list to store grouped elements
grouped_list_mobile = []

# Iterate through the input list
for i in range(0, len(data_mobile), 3):
    group = {}
    for j in range(3):
        if 'Image' in data_mobile[i + j]:
            group['Image'] = data_mobile[i + j]['Image']
            group['SSIM Value'] = data_mobile[i + j]['SSIM Value']
        if 'Values' in data_mobile[i + j]:
            if 'Values' not in group:
                group['Values'] = []
            group['Values'].extend(data_mobile[i + j]['Values'])
    grouped_list_mobile.append(group)

# Calculate averages of chi score
chi_values_list_mobile = []
for i in range (len(grouped_list_mobile)):
    #chi_values.append(grouped_list[i]['Values'])
    average = sum(grouped_list_mobile[i]['Values']) / len(grouped_list_mobile[i]['Values'])
    #print("Average Chi Value:", average)
    # Normalization of chi score
    average = 1/(1+average)
    #print("Normalized Average Chi Value:", average)
    chi_values_list_mobile.append(average)

# Iterate through the list
ssim_value_list_mobile = []
for item in data_mobile:
    if 'SSIM Value' in item:
        ssim_value = item['SSIM Value']
        
        # Print or use the extracted data as needed
        #print("SSIM Value:", ssim_value)
        ssim_value_list_mobile.append(ssim_value)

# Calculate an average combined score
average_score_mobile =[]
for i in range(len(chi_values_list_mobile)):
    average_score_mobile.append( (chi_values_list_mobile[i] + ssim_value_list_mobile[i])/2);
    #print(average_score_mobile[i])






# Open the text file for reading  
with open(filename_vgg, 'r') as file:
    entry = {}
    for line in file:
        line = line.strip()
        if line.startswith('Image and Style:'):
            entry['Image'] = line.split(':')[1].strip()
        elif line.startswith('SSIM Value:'):
            entry['SSIM Value'] = float(line.split(':')[1].strip())
        else:
            entry['Values'] = [float(value) for value in line.split()]
            data_vgg.append(entry)
            entry = {}

# Initialize a new list to store grouped elements
grouped_list_vgg = []

# Iterate through the input list
for i in range(0, len(data_vgg), 3):
    group = {}
    for j in range(3):
        if 'Image' in data_vgg[i + j]:
            group['Image'] = data_vgg[i + j]['Image']
            group['SSIM Value'] = data_vgg[i + j]['SSIM Value']
        if 'Values' in data_vgg[i + j]:
            if 'Values' not in group:
                group['Values'] = []
            group['Values'].extend(data_vgg[i + j]['Values'])
    grouped_list_vgg.append(group)

# Calculate averages of chi score
chi_values_list_vgg = []
for i in range (len(grouped_list_vgg)):
    #chi_values.append(grouped_list[i]['Values'])
    average = sum(grouped_list_vgg[i]['Values']) / len(grouped_list_vgg[i]['Values'])
    #print("Average Chi Value:", average)
    # Normalization of chi score
    average = 1/(1+average)
    #print("Normalized Average Chi Value:", average)
    chi_values_list_vgg.append(average)

# Iterate through the list
ssim_value_list_vgg = []
for item in data_vgg:
    if 'SSIM Value' in item:
        ssim_value = item['SSIM Value']
        
        # Print or use the extracted data as needed
        #print("SSIM Value:", ssim_value)
        ssim_value_list_vgg.append(ssim_value)

# Calculate an average combined score
average_score_vgg =[]
for i in range(len(chi_values_list_vgg)):
    average_score_vgg.append( (chi_values_list_vgg[i] + ssim_value_list_vgg[i])/2);
    #print(average_score_vgg[i])



# Final evaluation, which architecture works best
First_Places = []
Second_Places = []
for i in range(len(image_strings)):
    value1 = average_score_vgg[i]
    value2 = average_score_mobile[i]
    value3 = average_score_efficient[i]

    if value1 > value2 and value1 > value3:
        print("For Image '"+image_strings[i]+f"' VGG has the largest score :{value1}.")
        First_Places.append("VGG")
        if value2 > value3:
            print("For Image '"+image_strings[i]+f"' Mobile has the second largest score :{value2}.")
            print()
            Second_Places.append("Mobile")
        else:
            print("For Image '"+image_strings[i]+f"' Efficient has the second largest score :{value3}.")
            print()
            Second_Places.append("Efficient")
            
    elif value2 > value1 and value2 > value3:
        print("For Image '"+image_strings[i]+f"' Mobile has the largest score :{value2}.")
        First_Places.append("Mobile")
        if value1 > value3:
            print("For Image '"+image_strings[i]+f"' VGG has the second largest score :{value1}.")
            print()
            Second_Places.append("VGG")
        else:
            print("For Image '"+image_strings[i]+f"' Efficient has the second largest score :{value3}.")
            print()
            Second_Places.append("Efficient")
    else:
        print("For Image '"+image_strings[i]+f"' Efficient has the largest score :{value3}.")
        First_Places.append("Efficient")
        if value1 > value2:
            print("For Image '"+image_strings[i]+f"' VGG has the second largest score :{value1}.")
            print()
            Second_Places.append("VGG")
        else:
            print("For Image '"+image_strings[i]+f"' Mobile has the second largest score :{value2}.")
            print()
            Second_Places.append("Mobile")
print(First_Places)
```
